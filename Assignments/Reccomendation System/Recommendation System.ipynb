```python
#importing necessary packages
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
#importing dataset
data = pd.read_csv("book2.csv",encoding = "latin1")
```

```python
data
```

```python
data_1 = data.iloc[:,1:]
```

```python
#renaming the columns
data_1 = data_1.rename({'User.ID':'UserID','Book.Title':'BookTitle','Book.Rating':'BookRating'},axis = 1)
```

```python
data_1
```

```python
#sorting the values by userid
data_1.sort_values(by="UserID")
```

```python
data_1.columns
```

```python
#unique values_counts in UserID
len(data_1.UserID.unique())
```

```python
#unique values_counts in BookTitle
len(data_1.BookTitle.unique())
```

```python
#convrting data from long format to wide format using pivot_table
books = data_1.pivot_table(index='UserID', columns='BookTitle', values='BookRating').reset_index(drop=True)
```

```python
books
```

```python
#replacing the index by uniqvalue of userid
books.index = data_1['UserID'].unique()
```

```python
books
```

```python
#missing values
books.fillna(0, inplace = True)
```

```python
books
```

```python
#using cosin similaritie to find simillarities between two users which is user to user because we have to recommend item to user
user_sim = 1 - pairwise_distances( books.values,metric='cosine')
```

```python
user_sim
```

```python
#storing values in data frame
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df
```

```python
#assigning unique userid to rows and columns
user_sim_df.index = data_1['UserID'].unique()
user_sim_df.columns = data_1['UserID'].unique()
```

```python
#filling digonal values
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:3, 0:6]
```
