```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_csv('Company_Data.csv')
data.head()
```

```python
data2 = data.copy()
data2.head()
```

```python
data2['US'] = data2['US'].map({'Yes':1,'No': 0})
data2['Urban'] = data2['Urban'].map({'Yes':1,'No': 0})
data2['ShelveLoc'] = data2['ShelveLoc'].map({'Bad':0,'Good':2,'Medium':1})
```

```python
data2
```

```python
data2['sales'] = 'Minor'
data2.loc[data2['Sales'] >= 7.49,'sales'] = 'Major'
data2.drop(['Sales'],axis = 1,inplace = True)
```

```python
data2.head()
```

```python
X = data2.iloc[:,0:10]
X
```

```python
Y = data2.iloc[:,10]
Y
```

```python
X.shape,Y.shape
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.4,random_state = 3)
```

```python
RD = RandomForestClassifier(n_estimators=100,max_features=3)
RD.fit(x_train,y_train)
```

```python
predict = RD.predict(x_test)
np.mean(predict==y_test)*100 
```

```python
predict
```

```python
df_t = pd.DataFrame({'Actual':y_test, 'Predicted':predict})
df_t 
```

```python
i = pd.DataFrame({'feature': list(x_train.columns),
                   'importance': RD.feature_importances_}).\
                    sort_values('importance', ascending = False)
i
```
