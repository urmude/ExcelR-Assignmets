```python
#importing necessary packages
import pandas as pd 
import numpy as np
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB
from sklearn import preprocessing
```

```python
#importing dataset
salary_train = pd.read_csv('SalaryData_Train.csv')
salary_train
```

```python
# importing test dataset
salary_test = pd.read_csv('SalaryData_Test.csv')
salary_test
```

```python
salary_train.columns
```

```python
salary_test.columns
```

```python
string_columns=['workclass','education','maritalstatus','occupation','relationship','race','sex','native']
```

```python
#label encodeing
label_encoder=preprocessing.LabelEncoder()
for i in string_columns:
    salary_train[i]=label_encoder.fit_transform(salary_train[i])
    salary_test[i]=label_encoder.fit_transform(salary_test[i])
```

```python
#splitting the data into train1,train2
col_names=list(salary_train.columns)
train_1=salary_train[col_names[0:13]]
train_2=salary_train[col_names[13]]
test_1=salary_test[col_names[0:13]]
test_2=salary_test[col_names[13]]
```

```python
#gaussian NB
Gmodel=GaussianNB()
train_pred_gau=Gmodel.fit(train_1,train_2).predict(train_1)
```

```python
#fitting on train data and predicting on test data
test_pred_gau=Gmodel.fit(train_1,train_2).predict(test_1)
```

```python
#accuracy score on train data using gaussian NB
train_acc_gau=np.mean(train_pred_gau==train_2)
train_acc_gau
```

```python
#accuracy score  on test data using gaussian NB
test_acc_gau=np.mean(test_pred_gau==test_2)
test_acc_gau
```

```python
#confusion matrix for gaussian model 
from  sklearn.metrics import confusion_matrix,classification_report
cfm = confusion_matrix(test_2,test_pred_gau)
print(cfm)
```

```python
# classification_report matrix for 
clf = classification_report(test_2,test_pred_gau)
print(clf)
```

```python
#creating modle using multinomial NB
Mmodel=MultinomialNB()
train_pred_multi=Mmodel.fit(train_1,train_2).predict(train_1)
```

```python
#fitting on train data and predict on test data
test_pred_multi=Mmodel.fit(train_1,train_2).predict(test_1)
```

```python
# accuaracy of train data using multinomial NB
train_acc_multi=np.mean(train_pred_multi==train_2)
train_acc_multi
```

```python
# accuracy of test data using multinomial NB
test_acc_multi=np.mean(test_pred_multi==test_2)
test_acc_multi
```

```python
# confusion matrix for multinomial model 
CFM = confusion_matrix(test_2,test_pred_multi)
CFM
```

```python
# classification report for multinomial NB
CLF = classification_report(test_2,test_pred_multi)
print(CLF)
```
